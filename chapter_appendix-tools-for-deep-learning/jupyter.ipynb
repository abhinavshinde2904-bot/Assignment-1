```python
# i) Integer
num = 25
print("Integer value:", num)
print("Type:", type(num))


# ii) Float
num = 13.75
print("Float value:", num)
print("Type:", type(num))


# iii) String
text = "abhinavshinde"
print("Value:", text)
print("Type:", type(text))


# iv) Boolean
is_student = True
print("Value:", is_student)
print("Type:", type(is_student))


# v) Dictionary
i = {"name": "abhinav"}
print("Value:", i)
print("Type:", type(i))


# vi) List
fruits = ["Apple", "Banana", "Mango"]
print("List value:", fruits)
print("Type:", type(fruits))


# vii) Tuple
numbers = (10, 20, 30)
print("Tuple value:", numbers)
print("Type:", type(numbers))


# viii) Set
h = {"abhinav"}
print("Value:", h)
print("Type:", type(h))
```
