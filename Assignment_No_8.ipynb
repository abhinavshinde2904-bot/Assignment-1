```python
# 1. Write a Python script to sort (ascending and descending) a dictionary by value

d = {'a': 3, 'b': 1, 'c': 2}

asc = dict(sorted(d.items(), key=lambda item: item[1]))
desc = dict(sorted(d.items(), key=lambda item: item[1], reverse=True))

print("Ascending:", asc)
print("Descending:", desc)
```

```python
# 2. Write a Python script to check if a given key already exists in a dictionary

d = {'name': 'Abhinav', 'age': 18}

key = 'age'

if key in d:
    print("Key exists")
else:
    print("Key does not exist")
```

```python
# 3. Write a Python script to merge two Python dictionaries

d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}

d1.update(d2)

print("Merged Dictionary:", d1)
```
