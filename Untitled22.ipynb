```python
my_list = [10, 20, 30, 40]

# Update
my_list[1] = 25
print("After update:", my_list)

# Append
my_list.append(50)
print("After append:", my_list)

# Insert
my_list.insert(2, 15)
print("After insert:", my_list)

# Delete using remove()
my_list.remove(30)
print("After remove:", my_list)

# Delete using pop()
my_list.pop()
print("After pop:", my_list)

# Delete using del
del my_list[0]
print("After del:", my_list)
```
